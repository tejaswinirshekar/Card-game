### Programming Challenge: Card Game

### Task 1: Create a shuffled deck of cards

##### In task 1, I am defining the deck of cards. Each card shows a number from 1 to 10, each number is in the deck four times for a total of 40 cards. I am using "random" package to shuffle the cards and the same is displayed in the output.

```python
import random

class CardDeck:
    def __init__(self):
        self.cards = []
        self.display_initial_cards()
        self.shuffle_cards()

    def display_initial_cards(self):
        self.cards = [num for num in range(1, 11)] * 4
        print("Initial deck of cards before shuffling:", self.cards)

    def shuffle_cards(self):
        random.shuffle(self.cards)

# Create a CardDeck instance
deck = CardDeck()

# Print the shuffled deck of cards
print("Shuffled Deck of Cards:", deck.cards)
```

### Task 2: : Draw cards
##### In task 2, it is shown that how the player is drawing the card and adding it to the discarded pile. You can try the below test cases by modifying in the "simulate" part as shown below.

Case 1: Drawing the card from the "draw pile".

Case 2: Drawing the card from the "draw pile" until atleast 1 card is present.

player.drawCard = [6]

Case 3: Drawing the card from the "discarded pile" after shuflling it, when the "draw pile" is empty.

player.drawCard = []
player.discardCard = [1, 3, 5]


```python
class Player:
    def __init__(self):
        self.drawCard = []
        self.discardCard = []

    def draw_cards(self):
        if len(self.drawCard) == 0:
            print("Draw pile is empty. Using shuffled discarded pile as draw pile.")
            if len(self.discardCard) == 0:
                print("Nothing to draw from, game over.")
                return 0
            random.shuffle(self.discardCard)
            self.drawCard = self.discardCard
            self.discardCard = []
            return self.draw_cards()  # Recursive call to draw_cards to draw from the new draw pile
        card = self.drawCard.pop()
        return card

    def discard_card_set(self, card):
        if type(card) == list:
            self.discardCard.extend(card)
        else:
            self.discardCard.append(card)
        return self.discardCard

# Create a Player instance
player = Player()

##Simulate

# Simulate an empty draw pile and add some cards to the discarded pile
player.drawCard = [2, 4, 6]
player.discardCard = [1, 3, 5]

# Draw a card (should use the shuffled discarded pile)
print("Drawn Card by Player:", player.draw_cards())

```

### Task 3: Playing a turn 

##### In task 3, the game starts, if you execute the below function, the 2 players gets the random 20 cards each and loop starts executing until there is a winner.

```python
def Game():
    # Start Game
    cardDeck = CardDeck()
    cardDeck.shuffle_cards()
    print("Initial shuffled deck of cards - " + str(cardDeck.cards))
    drawnCardForPlayer1 = cardDeck.cards[0:20]
    drawnCardForPlayer2 = cardDeck.cards[20:40]
    print("Start Game with cards (drawnCardForPlay1) - " + str(drawnCardForPlayer1))
    print("Start Game with cards (drawnCardForPlay2) - " + str(drawnCardForPlayer2))

    # Initialize the Players
    player1 = Player()
    player1.drawCard = drawnCardForPlayer1
    player2 = Player()
    player2.drawCard = drawnCardForPlayer2

    tied_rounds = 0
    collected_cards = []
    turn = 0

    while True:
        turn += 1
        print("\nPlaying round:", turn)
        print("Draw Pile for Player 1:", player1.drawCard, "(Total:", len(player1.drawCard), "cards)")
        print("Draw Pile for Player 2:", player2.drawCard, "(Total:", len(player2.drawCard), "cards)")

        # Task 2: Draw Cards
        cardPlayer1 = player1.draw_cards()
        if cardPlayer1 != 0:
            print("Card drawn by Player 1:", cardPlayer1)
        else:
            if len(player1.discardCard) != 0:
                random.shuffle(player1.discardCard)
                player1.drawCard = player1.discardCard
                print("Discarded deck for Player 1:", player1.drawCard)
                cardPlayer1 = player1.draw_cards()
                print("Card drawn by Player 1 from discarded deck:", cardPlayer1)
            else:
                print("GAME OVER, Player 2 WINS")
                break

        cardPlayer2 = player2.draw_cards()
        if cardPlayer2 != 0:
            print("Card drawn by Player 2:", cardPlayer2)
        else:
            if len(player2.discardCard) != 0:
                random.shuffle(player2.discardCard)
                player2.drawCard = player2.discardCard
                print("Discarded deck for Player 2:", player2.drawCard)
                cardPlayer2 = player2.draw_cards()
                print("Card drawn by Player 2 from discarded deck:", cardPlayer2)
            else:
                print("GAME OVER, Player 1 WINS")
                break

        # Task 3: Playing a turn
        if cardPlayer1 == cardPlayer2:
            tied_rounds += 1
            collected_cards.append(cardPlayer1)
            collected_cards.append(cardPlayer2)
            print("Both players discard their cards.")

        if cardPlayer1 > cardPlayer2:
            if tied_rounds > 0:
                player1.discard_card_set(collected_cards)
                tied_rounds = 0
                collected_cards = []

            player1.discard_card_set(cardPlayer1)
            player1.discard_card_set(cardPlayer2)
            print("Player 1 wins this round!")
                
        if cardPlayer2 > cardPlayer1:
            if tied_rounds > 0:
                player2.discard_card_set(collected_cards)
                tied_rounds = 0
                collected_cards = []

            player2.discard_card_set(cardPlayer1)
            player2.discard_card_set(cardPlayer2)
            print("Player 2 wins this round!")

        discardedCardPlayer1 = player1.discardCard
        ("Draw Pile for Player 1:", player1.drawCard, "(Total:", len(player1.drawCard), "cards)")
        print ("Discarded Card Player 1:", discardedCardPlayer1, "(Total:", len(discardedCardPlayer1), "cards)")
        discardedCardPlayer2 = player2.discardCard
        print ("Discarded Card Player 2:", discardedCardPlayer2, "(Total:", len(discardedCardPlayer2), "cards)")

Game()

```
